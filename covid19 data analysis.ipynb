```python
#import necessary modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#open covid19 file
covid_confirmed_data=pd.read_csv('data_analysis/covid19.csv')
covid_confirmed_data.head()
```

```python
#drop unnecessary columns
covid_confirmed_data.drop(["Lat","Long"],axis=1,inplace=True)
covid_confirmed_data
```

```python
#sum the values and use groupby to put duplicates together into one
covid_data=covid_confirmed_data.groupby("Country/Region").sum()
covid_data
```

```python
covid_china=covid_data.loc['China'].plot()
```

```python
covid_data.loc['China'].head(3).plot()
#Notice increase in cases in a 24 hour period
#day1=100(650-550)
```

```python
#compare any 3 countries
covid_data.loc['China'].plot()
covid_data.loc['Italy'].plot()
covid_data.loc['Spain'].plot()
plt.legend()
```

```python
#calculate first derivative
covid_data.loc['China'].diff().plot()
```

```python
#find maximum
covid_data.loc['China'].diff().max()
```

```python
#find max for all countries and store it in separate column
countries=list(covid_data.index)
#countries
max_infection_rate=[]
for country in countries:
    max_infection_rate.append(covid_data.loc[country].diff().max())
#max_infection_rate
covid_data['max_infection_rate']=max_infection_rate
covid_data
```

```python
#keep necessary columns and remove the rest
covid_final_data=pd.DataFrame(covid_data["max_infection_rate"])
covid_final_data
```

```python
#covid_final_data.plot()
#import world_happiness_report file
happiness_report=pd.read_csv('data_analysis/world_happiness_report.csv')
happiness_report
```

```python
#keep only necessary columns
happiness_report.drop(['Overall rank','Score','Generosity','Perceptions of corruption'],axis=1,inplace=True)
happiness_report
```

```python
#reset_index to make it easier for merging two files
happiness_report.set_index('Country or region')
happiness_report.reset_index()
happiness_report.rename(columns={'Country or region':'Country/Region'},inplace=True)
happiness_report=happiness_report.set_index('Country/Region')
happiness_report
```

```python
covid_final_data.shape
```

```python
happiness_report.shape
```

```python
#merge the files 
data=covid_final_data.join(happiness_report,how='inner')#since the files are of diff. shapes,use inner join
data
```
